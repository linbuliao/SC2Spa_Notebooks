```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
BM = pd.read_excel('Benchmarking/Benchmark.xlsx', skiprows=1, usecols = [2,3,4,5,6], index_col = 0)[:-4]
BM1 = BM[:4]
BM2 = BM[4:]
```

```python
BM1
```

```python
def Barplot(BM1, BM2, BM1_name, BM2_name):
    
    ind = np.arange(2)
    colors = ['C3', 'C4', 'C1', 'C0']
    width = 0.15

    for column in BM2.columns:
        print(column)
        fig = plt.figure()
        ax = fig.add_axes([0,0,1,1])
        for i in range(BM2.shape[0]):
            ax.bar(ind + width * i, [BM1.iloc[i][column], BM2.iloc[i][column]],
                   fill = True, color = colors[i], width = width-0.04)
        ax.set_xticks(ind + 1.5*width)
        ax.set_xticklabels([BM1_name, BM2_name], fontsize = 18)
        ax.legend(labels=BM2.index.tolist())
        plt.title(column, fontsize = 24)
        plt.savefig('Benchmarking/'+column+'.png', bbox_inches = 'tight')
        plt.show()
```

```python
ind = np.arange(2)
colors = ['C3', 'C4', 'C1', 'C0']
width = 0.15

for column in BM2.columns:
    print(column)
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    for i in range(BM2.shape[0]):
        ax.bar(ind + width * i, [BM1.iloc[i][column], BM2.iloc[i][column]],
               fill = True, color = colors[i], width = width-0.04)
    ax.set_xticks(ind + 1.5*width)
    ax.set_xticklabels(['Hippocampus', 'Embryo'], fontsize = 18)
    ax.legend(labels=BM2.index.tolist())
    plt.title(column, fontsize = 24)
    plt.savefig('Benchmarking/'+column+'.png', bbox_inches = 'tight')
    plt.show()
```
