```python
import numpy as np
import pandas as pd
import anndata as ad
import random

from scipy import stats
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import LogisticRegression

import keras
from keras.models import Sequential, Model, load_model
from keras.regularizers import l2
from keras.layers import Dense, Input, BatchNormalization, Dropout, Convolution1D, Activation, Concatenate, concatenate
from keras.layers.pooling import AveragePooling1D, GlobalAveragePooling1D, MaxPooling1D, MaxPool1D
from keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint

import keras.backend as K

import os
import gc
import pickle
import time

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pct_offset = 0

def rmse(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true), axis=-1))
def calc_rmse(y_true, y_pred):
    return np.sqrt(np.square(y_true - y_pred).mean())

loss = rmse
loss_name = 'rmse'

def MinMaxNorm(Y):
    return (Y-Y.min(axis = 0))/(Y.max(axis = 0)-Y.min(axis = 0))/(1+pct_offset)

def ReMMNorm(Y, Y_pred):
    return (Y_pred*(Y.max(axis=0)-Y.min(axis=0)*(1+pct_offset))+Y.min(axis = 0))

def PolarTrans(Y):
    
    R = np.sqrt(np.square(Y[:,0]) + np.square(Y[:,1]))
    Theta = np.arctan(Y[:,1]/Y[:,0])
    RTheta = np.concatenate([R.reshape(-1,1), Theta.reshape(-1,1)], axis = 1)
    
    return RTheta

def RePolarTrans(RTheta):
    x = RTheta[:,0] * np.cos(RTheta[:,1])
    y = RTheta[:,0] * np.sin(RTheta[:,1])
    Y = np.concatenate([x.reshape(-1,1), y.reshape(-1,1)], axis = 1)
    return Y

def batch_generator(X, y, sample_index, batch_size, shuffle):
    number_of_batches = int(np.ceil(len(sample_index)/batch_size))
    counter = 0

    if shuffle:
        np.random.shuffle(sample_index)
    while True:
        batch_index = sample_index[batch_size*counter:batch_size*(counter+1)]
        X_batch = X[batch_index, :]
        y_batch = y[batch_index, :]
        counter += 1
        yield X_batch, y_batch
        if (counter == number_of_batches):
            if shuffle:
                np.random.shuffle(sample_index)
            counter = 0

def FCNN(input_shape, out_shape,l2_reg = 1e-4, dropout = 0.15, nodes = [2048, 1024, 512]):
    
    X_input = Input(input_shape)
    
    X = X_input
    
    for node in nodes:
        X = Dropout(rate = dropout)(X)
        X = Dense(node, kernel_initializer = 'glorot_normal', activation = 'relu', kernel_regularizer = l2(l2_reg))(X)
        X = BatchNormalization(axis = 1)(X)
    
    X = Dense(out_shape, kernel_initializer = 'glorot_normal', activation = 'sigmoid')(X)
    
    model = Model(inputs = X_input, outputs = X, name = 'FCNN')
    
    return model

def DatasetMapping(adata_SR, adata_target, l2_reg = 1e-4, dropout = 0.15,
                   epoch = 500, batch_size = 1024, nodes = [4096, 1024, 256, 64, 16, 4]):
    
    #Extract information
    X_SR = adata_SR.X
    Y_SR = adata_SR.obs[['xcoord', 'ycoord']].values
    #X_target = adata_target.X
    
    #Coordinate Transformation
    RTheta_SR = PolarTrans(Y_SR)
    RThetaNorm_SR = MinMaxNorm(RTheta_SR)
    
    length = RThetaNorm_SR.shape[1]

    earlystopper = EarlyStopping(monitor=loss_name, patience=50, verbose=1)

    # Set a learning rate annealer
    learning_rate_reduction = ReduceLROnPlateau(monitor=loss_name, 
                                                patience=3, 
                                                verbose=1, 
                                                factor=0.5, 
                                                min_lr=1e-9)

    model = FCNN((X_SR.shape[1],), length, l2_reg = l2_reg, dropout = dropout, nodes = nodes)
    model.compile(optimizer = 'adam', loss = loss, metrics = [loss])

    model.fit_generator(generator = batch_generator(X_SR,\
                                                    RThetaNorm_SR,\
                                                    np.arange(X_SR.shape[0]),\
                                                    batch_size, True),\
                        epochs = epoch,\
                        steps_per_epoch = X_SR.shape[0] / batch_size,\
                        callbacks = [learning_rate_reduction, earlystopper])

    Target_pred_RTN = model.predict(adata_target.X)
    Target_pred_Y = RePolarTrans(ReMMNorm(RTheta_SR, Target_pred_RTN))
    
    return model, Target_pred_Y
```

```python
def Visualize2(adata, coord, figsize = (7, 6), select = None, c_name = 'leiden',
               xlim = [650, 5750], ylim = [650, 5750], title = 'SILE'):
    
    save = 'figures_transfer/'
    if not os.path.exists(save):
        os.makedirs(save)

    if(select!=None):
        CTs = adata.obs[c_name].value_counts()
        is_doub = CTs.index.str.contains('_')

        SelectedCTs = CTs[~is_doub].index.tolist()[select : (select + 12)]
        #SelectedCTs = CTs.index.tolist()[:8]

        t = adata
        Select1 = t.obs[c_name].apply(lambda x: x in SelectedCTs)
        Select = Select1.astype(bool)
        adata = t[Select]
        coord = coord[Select]    

    color_dic = {}
    for i, ct in enumerate(list(adata.obs[c_name].unique())):
        color_dic[ct] = i

    color = list(map(lambda x: color_dic[x], adata.obs[c_name]))

    plt.figure(figsize =figsize)
    scatter = plt.scatter(coord[:, 0],\
                coord[:, 1],\
                s = 3,\
                c = color,\
                cmap = plt.get_cmap('Paired'))
    plt.grid(False)
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.title(title)

    plt.legend(handles=scatter.legend_elements()[0],
           labels = list(adata.obs[c_name].unique()),
           bbox_to_anchor=(1, 0.9))

    plt.savefig(save + title + '.png', bbox_inches='tight')
    plt.show()
```
